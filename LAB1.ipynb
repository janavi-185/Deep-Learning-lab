Where:

```python
# IMPORTS
import numpy as np
```

### Exercise 1: Scalars and Vectors
#### Objective: Understand how scalars and vectors are represented in NumPy.
##### • Scalar: A single numerical value (e.g., bias)
##### • Vector: A one-dimensional array representing a single data sample

```python
bias = 0.5
x = np.array([1.0, 2.0, 3.0])
w = np.array([0.2, 0.4, 0.6])
```

```python
print("Bias (scalar):", bias)
print("Input Vector:", x)
print("Weight Vector:", w)
```



### Exercise 2: Dot Product
#### Objective: Compute the dot product between input and weight vectors.
### What is a Dot Product?
The dot product (also called inner product or scalar product) is:
This is a weighted sum of inputs where each input is multiplied by its corresponding weight, then all products are summed.

```python
dot_product = np.dot(w, x)
print("Dot Product (w · x):", dot_product)
```



### Exercise 3: Linear Combination
#### Objective: Compute the linear output of a neuron by adding bias to dot product.

```python
z = dot_product + bias
print("Linear Output (z):", z)
```



### Exercise 4: Step Activation Function
#### Objective: Convert linear output into a binary decision using an activation function.
##### An activation function introduces non-linearity to the model. A simple step function is:


```python
def step_function(z):
    return 1 if z >= 0 else 0

output = step_function(z)
print("Perceptron Output:", output)
```



### Exercise 5: Matrix as Batch of Samples
#### Objective: Understand batch processing with matrices where each row is a data sample.



```python
X = np.array([
    [1.0, 2.0, 3.0],
    [0.5, 1.5, 2.5],
    [2.0, 1.0, 0.5]
])

print("Input Matrix (Batch of Samples):")
print(X)
```



### Exercise 6: Batch Dot Product
#### Objective: Apply dot product to multiple samples simultaneously and add bias.


```python
batch_output = np.dot(X, w) + bias
print("Batch Linear Output:")
print(batch_output)
```

## Post-Lab Reflection Questions

### 1. What role do weights play in the dot product?
Weights determine the importance of each input feature in the model.  
In the dot product, each input value is multiplied by its corresponding weight, allowing the model to emphasize more relevant features and reduce the impact of less important ones.

### 2. How does bias affect the decision boundary?
Bias shifts the decision boundary of the model.  
Without bias, the decision boundary must pass through the origin, but adding bias allows the model to fit data more flexibly by shifting the boundary in feature space.

### 3. Why is dot product fundamental to neural networks?
The dot product computes the weighted sum of inputs, which is the core operation performed by a neuron.  
Every neural network layer relies on dot products to combine inputs and weights before applying an activation function.

### 4. How does batch computation help in training deep learning models?
Batch computation allows the model to process multiple data samples simultaneously.  
This improves computational efficiency, makes better use of hardware like GPUs, and leads to faster and more stable training.
